```python
import pandas as pd
import requests
import tarfile
from os import path
```

```python
fname = 'airline_2m.tar.gz'
url = 'https://dax-cdn.cdn.appdomain.cloud/dax-airline/1.0.1/' + fname
r = requests.get(url)
open(fname , 'wb').write(r.content)
```

```python
tar = tarfile.open(fname)
tar.extractall()
tar.close()
```

```python
data_path = "airline_2m.csv"
path.exists(data_path)
```

```python
df = pd.read_csv(data_path, encoding = "ISO-8859-1", dtype={'Div1Airport': str, 'Div1TailNum': str, 'Div2Airport': str, 'Div2TailNum': str})
df.head()
```

```python
#Load raw airportsdata csv
url = 'https://raw.githubusercontent.com/mborsetti/airportsdata/main/airportsdata/airports.csv'
airports_data = pd.read_csv(url, index_col=0)

print(airports_data['iata'])
```

```python
df1 = df.merge(airports_data, left_on='Origin', right_on='iata')
print(df1.head(5))
```

```python
df1.rename(columns = {'lat':'Origin Latitude', 'lon':'Origin Longitude'})
```

```python
df2 = df1.merge(airports_data, left_on='Dest', right_on='iata')
print(df2.head())
```

```python
df2 = df2.rename(columns={'lat_x':'Origin Latitude', 'lon_x':'Origin Longitude', 'lat_y':'Destination Latitude', 'lon_y':'Destination Longitude'})
```
